```python
import nltk
import re
import string

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download NLTK data
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)

stop_words = set(stopwords.words('english'))

# ----------------------------------------
# Training Data
# ----------------------------------------

texts = [
    # Positive
    "I am very happy",
    "This is amazing",
    "I love this product",
    "You are doing great",
    "I feel fantastic today",
    "That was wonderful",
    "This makes me smile",
    "I am excited about this",
    "Everything is awesome",
    "I am proud of you",

    # Negative
    "I am very sad",
    "This is terrible",
    "I hate this",
    "I feel disappointed",
    "This is the worst",
    "I am upset",
    "That was horrible",
    "I feel angry",
    "This makes me frustrated",
    "I am not happy"
]

labels = [1]*10 + [0]*10   # 1 = Positive, 0 = Negative

# ----------------------------------------
# Text Preprocessing
# ----------------------------------------

def preprocess(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = word_tokenize(text)
    tokens = [w for w in tokens if w not in stop_words]
    return " ".join(tokens)

processed_texts = [preprocess(t) for t in texts]

# ----------------------------------------
# TF-IDF Vectorization
# ----------------------------------------

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(processed_texts)

# ----------------------------------------
# Train Model
# ----------------------------------------

model = LogisticRegression()
model.fit(X, labels)

# ----------------------------------------
# Film-Based Feedback Data
# ----------------------------------------

positive_movies = [
    "Watch a feel-good movie like *3 Idiots* ðŸŽ‰",
    "You may enjoy *The Pursuit of Happyness* ðŸ˜Š",
    "Try a comedy like *PK* ðŸ˜„"
]

negative_movies = [
    "You might relate to *Taare Zameen Par* ðŸŽ¬",
    "A calm movie suggestion: *Dear Zindagi* ðŸŒ±",
    "Try something emotional like *Kal Ho Naa Ho* ðŸ’™"
]

# ----------------------------------------
# Model Feedback
# ----------------------------------------

predictions = model.predict(X)
print("Model Accuracy:", accuracy_score(labels, predictions))
print("\nðŸ¤– Film Feedback Chatbot Ready! Type 'exit' to stop.\n")

# ----------------------------------------
# Interactive Chat Loop
# ----------------------------------------

import random

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Chatbot: Goodbye! ðŸŽ¬")
        break

    processed_input = preprocess(user_input)
    input_vector = vectorizer.transform([processed_input])
    prediction = model.predict(input_vector)[0]

    if prediction == 1:
        print("Chatbot: You seem happy! ðŸ˜Š")
        print("Film Feedback:", random.choice(positive_movies))
    else:
        print("Chatbot: You seem low ðŸ˜Ÿ")
        print("Film Feedback:", random.choice(negative_movies))
```
